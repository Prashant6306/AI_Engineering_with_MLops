### Convert Set Elements as Keys with Default Values

```python
set1={"a","b","c"}
dict1=dict()
count=10
for key in set1:
    dict1[key]=count
    count=count+10
print(dict1)
```

### Aliasing

```python
original_dict = {1: 'a', 2: 'b', 3: 'c'}
alias_dict = original_dict
print(id(original_dict))  # 1024113
print(id(alias_dict))     # 1024113  ← Same!
```

### Cloning

```python
original_dict = {1: 'a', 2: 'b', 3: 'c'}
cloned_dict = original_dict.copy()
print(id(original_dict))  # 100123
print(id(cloned_dict))    # 100124
```

### Searching

```python
my_dict = {1: 'a', 2: 'b', 3: 'c'}

if (3 in my_dict):
    print("3 is present in dict")
else:
    print("3 is not present in dict")
```

### Identity — 'is' Operator

```python
dict1 = {1: 'a', 2: 'b', 3: 'c'}
dict2 = dict1              # aliasing
print(dict1 is dict2)      # True
print(dict2 is dict1)      # True
```

```python
dict3 = {1: 'a', 2: 'b', 3: 'c'}
dict4 = dict3.copy()       # cloning
print(dict3 is dict4)      # False  ← different objects
print(dict4 is dict3)      # False
```

### Comparison

```python
dict_a = {1: 'a', 2: 'b', 3: 'c'}
dict_b = {1: 'a', 2: 'b', 3: 'c'}
dict_c = {1: 'x', 2: 'y'}

print(dict_a == dict_b)    # True  ← same content
print(dict_a == dict_c)    # False ← different content
print(dict_a != dict_c)    # True
```

### Emptiness
