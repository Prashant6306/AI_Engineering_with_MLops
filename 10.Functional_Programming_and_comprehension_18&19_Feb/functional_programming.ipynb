### Lambda Functions

```python
##### Syntax: lambda arguments : expression
```

```python
def multiply(p,q):
    return p*q

multiply=multiply(6,7)
print(multiply)
```

```python
multiply = lambda p, q: p * q
print(multiply(6, 7))
```

### Check If a Number is Positive

```python
def is_positive(n):
    return (n>0)


print(is_positive(-3))
```

```python
print(is_positive(3))
```

```python
##### Syntax: lambda arguments : expression

is_positive=lambda q:(q>0)
print(is_positive(-3))
print(is_positive(3))
```

#### Concatenate Two Strings

```python
join_words = lambda a, b, c: a.lower() + " " + b.lower() + " "+ c.lower()
print(join_words("Hello", "Python","Programming"))
```

```python
odd_no=lambda q: "odd" if q%2!=0 else "even"
print(odd_no(10))
```
