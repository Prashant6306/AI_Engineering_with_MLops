## List Comprehension

#### syntax : [ expression for item in iterable/sequence ]

```python
for i in range(1,6):
    print(i)
```

### Squares of numbers 1 to 5

```python
#[ expression for item in iterable/sequence ]
list1=[item**2 for item in range(1,6)]
```

```python
print(list1)
```

### cube of numbers 1 to 5

```python
list2=[item**3 for item in range(1,6)]
```

```python
print(list2)
```

```python
for i in range(1,21):
    print(i)
```

#Even numbers from 1 to 20

```python
#[ expression for item in iterable/sequence if condtion ]

list3=[item for item in range(1,21) if item%2==0]

#[1,2,3,---------20]
```

```python
print(list3)
```

```python
list1=[]
for item in range(1,21):
    if (item%2 ==0):
        list1.append(item)

print(list1)
```

```python
print(list3)
```

### Convert words to uppercase

```python
animals=['cat','dog','fish','bird']
```

```python
#[ expression for item in iterable/sequence if condtion ]

list2=[item.upper() for item in animals ]
```

```python
print(list2)
```

```python
list3=[item.title() for item in animals ]
```

```python
print(list3)
```

```python
animals=['cat','dog','fish','bird','big']
```

```python
list3=[item for item in animals if item.startswith('b')]
```

```python
print(list3)
```

## Dictionary Comprehension

```python
# syntax: { key_expression : value_expression for item in sequence/iterable if condition }
```

```python
list1=["a","b","c"]
```

```python
dict1={item:10 for item in list1 }
```

```python
print(dict1)
```

#### convert no to seq in this form -> {0:0, 1:1, 2:4, 3:9, 4:16}

```python
dict2={item:item**2 for item in range(0,5)}
```

```python
print(dict2)
```

#### animals=['cat','dog','elephant'] -> {'cat':3, 'dog':3, 'elephant':8}

```python
animals=['cat','dog','elephant']
```

```python
dict2={item:len(item) for item in animals}
```

```python
print(dict2)
```

### Student → Grade (pass only) ->{'Alice':82, 'Carol':91} 

```python
Students={'Alice':82,'Bob':45,'Carol':91}
```

```python
dict2={key:value for key,value in Students.items() if (value>=50)}
```
