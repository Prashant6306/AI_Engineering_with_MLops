```python
class Student:
    def __init__(self,name,roll_no):
        self.name=name
        self.roll_no=roll_no
        print("Constructor has been called")
        self.dispaly_info()

    def dispaly_info(self):
        print("name:",self.name)
        print("roll no:",self.roll_no)
        self.demo_fun()

    def demo_fun(self):
        print("demo function has been callled")

student1=Student("Ganesh",101)
student2=Student("Santosh",102)
print(student1.name)
student1.name="Mahesh"
print(student1.name)
print(student2.name)
```
